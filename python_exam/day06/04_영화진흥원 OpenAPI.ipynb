### 1) 필요한 패키지 가져오기

```python
import requests
import json
import datetime as dt
import pandas as pd
from pandas import DataFrame
from matplotlib import pyplot
```

### 2) API 키 준비하기 

```python
api_key = "62aa093993a32d3280249fc8417a88cb"
```

### 3) 세션 생성

```python
user_agent = "Mozilla/5.0 (Linux; Android 5.0; SM-G900P Build/LRX21T) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Mobile Safari/537.36"
session = requests.Session()
session.headers.update({'User-agent':user_agent, 'referer':None})
```

## #02. 특정 날짜에 대한 박스오피스 관람객 수

영화진흥위원회의 OpenAPI는 오늘을 기준으로 하루 전날 까지의 데이터를 제공한다.

### 1) 접속 주소 준비하기 

#### 변수가 치환될 주소 템플릿

```python
url_tpl = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key={key}&targetDt={date}"
```

#### 조회날짜 문자열 만들기
> 여기서는 어제 날짜

```python
today = dt.datetime.now()                     # 오늘 날짜 
delta = dt.timedelta(days = -1)               # 하루 전을 의미하는 timedelta 객체
yesterday = today + delta                     # 오늘날짜와 timedelta 연산
yesterday_str = yesterday.strftime("%Y%m%d")  # yyyymmdd 형식 문자열로 변환
yesterday_str
```

```python
# 최종 접속 주소 확인
api_url = url_tpl.format(key=api_key, date=yesterday_str)
api_url
```

### 3) API를 통한 JSON 데이터 가져오기 

#### OpenAPI를 통한 json 가져오기

```python
r = session.get(api_url)

if r.status_code != 200 : 
    print("Error")
    quit()

r.encoding = "utf-8"
r.text
```

#### json을 딕셔너리로 변환 시키자.

```python
dic = json.loads(r.text)
dic
```

###  딕셔너리를 데이터프레임으로 변환

```python
df = DataFrame(dic['boxOfficeResult']['dailyBoxOfficeList'])
df
```

### 데이터 전처리

```python
t_df = df.filter(['movieNm', 'audiCnt'])
t_df
```

### 영화제목을 인덱스로 지정

```python
name_dic = {}
for i, v in enumerate(list(t_df['movieNm'])):
    name_dic[i] =v

r_df = t_df.rename(index=name_dic, columns={'audiCnt' : '관람객'})
r_df.drop('movieNm',axis=1, inplace=True)
r_df
```

#### 관람객 자료형 확인

```python
r_df['관람객']
# object 이면 string 이다.
```

###  관람객 타입을 숫자형식으로 변환

```python
r_df['관람객'] = r_df['관람객'].apply(pd.to_numeric)
r_df['관람객']
```

### 시각화

```python
# 그래프 만들기
pyplot.rcParams["font.family"] = 'Malgun Gothic'
pyplot.rcParams["font.size"] = 18
pyplot.rcParams["figure.figsize"] = (32, 16)

# 전체 컬럼에 대한 시각화
r_df.plot.barh(rot=45)
pyplot.grid()
pyplot.title("%s 박스오피스 순위" % yesterday_str)
pyplot.legend()
pyplot.show()
pyplot.close()
```

## 3. 영화별 1주일 간의 관람객 수 변동 현황 시각화

> URL에 접속하는 연동 부분을 하나의 블록으로 묶고 날짜을 변경해 1주일간의 변화량을 확인한다.

### 1) 반복문을 이용해서 1주일치 데이터 수집하기

```python
# 반복을 수행하면서 생성되는 데이터프레임을 누적할 빈 데이터프레임 준비
c_df = DataFrame()

# 일주일 간의 기간을 위한 반복문
for i in range(-7, 0):
    today = dt.datetime.now()                     # 오늘 날짜 
    delta = dt.timedelta(days = i)               # 하루 전을 의미하는 timedelta 객체
    yesterday = today + delta                     # 오늘날짜와 timedelta 연산
    yesterday_str = yesterday.strftime("%Y%m%d")  # yyyymmdd 형식 문자열로 변환
    # print(yesterday_str)

    #### 최종 접속 주소 확인
    api_url = url_tpl.format(key=api_key, date=yesterday_str)
    
    ### API를 통한 JSON 데이터 가져오기 
    r = session.get(api_url)
    
    ## 오류 발생하면 멈추는 것이 아니라 다음날 반복 진행 
    if r.status_code != 200 : 
        print("Error")
        # quit()
        continue

    r.encoding = "utf-8"

    ## 데이터프레임으로 변환
    dic = json.loads(r.text)
    df = DataFrame(dic['boxOfficeResult']['dailyBoxOfficeList'])
    
    # 사용할 컬럼만 추출
    t_df = df.filter(['movieNm','audiCnt'])
    
    # 영화제목을 인덱스로 지정하기 
    name_dic = {}
    for i, v in enumerate(list(t_df['movieNm'])):
        name_dic[i] = v
    
    # 인덱스 이름변경, 컬럼이름 변경
    r_df = t_df.rename(index=name_dic, columns={'audiCnt' : yesterday_str})
    r_df.drop('movieNm', axis=1, inplace=True)
    # print(r_df)
    
    # 관람객 컬럼의 타입을  숫자로 변경
    r_df[yesterday_str] = r_df[yesterday_str].apply(pd.to_numeric)
    
    ### 영화이름이 동일한 데이터는 누적
    ### how값을 outer로 지정하여 중복되지 않는 인덱스명도 누락되지 않도록 한다.
    c_df = pd.merge(c_df, r_df, left_index=True, right_index=True, how="outer")
    
print(c_df)
```

#### 결측치를 0으로 대체

```python
f_df = c_df.fillna(0)
```

## 시각화

### 시간에 따른 데이터의 변화를 표현하기 위해서는 선 그래프(시계열 그래프) 사용

```python
# 그래프 만들기
pyplot.rcParams["font.family"] = 'Malgun Gothic'
pyplot.rcParams["font.size"] = 18
pyplot.rcParams["figure.figsize"] = (32, 16)

# 전체 컬럼에 대한 시각화
f_df.plot(rot=45)
pyplot.grid()
pyplot.title("일주일깐 영화별 관람객 수 변화")
pyplot.legend(bbox_to_anchor=(1,1))
pyplot.show()
pyplot.close()
```
