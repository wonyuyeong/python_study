````python
from pandas import ExcelFile
import folium
````

##  #02. 서울 초등학교 찾기

### 1) 데이터 수집

````python
xlsx = ExcelFile('school2019.xlsx')
df = xlsx.parse(xlsx.sheet_names[0])
df.head()
````

### 2) 사용할 필드만 추출

````python
df2 = df.filter(['학교명','학교급구분', '소재지도로명주소', '위도','경도'])
df2
````

### 3) 서울시의 초등학교만 추출

1. `학교급구분` 필드값이 **초등학교**
1. `소재지도로명주소`에  **서울** 이라는 단어가 포함되는 경우
    ```
    # DB의 LIKE 연산
    컬럼명.str.contains('검색어')
    ```

````python
# & = and 
df3 = df2[(df2["학교급구분"] == '초등학교') & (df2["소재지도로명주소"].str.contains('서울'))]
df3
````

## #03.데이터 시각화

### 1) 지도 객체 생성

````python
# zoom_start : 배율 (1-22)
map = folium.Map(location=[37.55258295334973, 126.9376976243557],zoom_start=18)
map
````

### 2) 데이터프레임의 행 수 만큼 반복하면서 마커 생성

````python
for i in df3.index:
    # 행 접근 => loc
    name = df3.loc[i,'학교명']
    lat = df3.loc[i, '위도']
    lng = df3.loc[i,'경도']
    
    # 마커 표시
    marker = folium.Marker([lat,lng], popup=name)
    marker.add_to(map)

map
````
