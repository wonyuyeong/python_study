### 조건에 따른 선택적인 값을 추가하기 

#### 조건이 참인 경우 A, 그렇지 않은 경우 B

```python
# 국어점수가 70이면 합격, 아니면 불합격을 국어결과 컬럼에 넣자
add_col['국어결과'] = numpy.where(add_col['국어']>=70, '합격', '불합격')
add_col
```

#### 영어점수를 90이상이면 A학점,  80이상이면 B학점, 70이상이면 C학점, 나머지는 F학점, 만약에 빈값(NaN) 이면 F학점

```python
conditions = [(add_col['영어'] >= 90),   # A 
              (add_col['영어'] >= 80),   # B
              (add_col['영어'] >= 70),   # C
              (add_col['영어'] < 70),    # F
              (numpy.isnan(add_col['영어']) == True)] # F
```

```python
# conditions 리스트 순서에 맞게 학점 부여
grade = ['A', 'B', 'C', 'F', 'F']
```

```python
add_col['영어학점'] = numpy.select(conditions, grade)
add_col
```

### 1) 행 단위 병합
#### 샘플 데이터 만들기

```python
df_top = DataFrame({'국어':[90,82], '수학':[81,76]}, index=['민철','철수'])
df_top
```

```python
df_middle = DataFrame({'국어':[70,62], '영어':[77,68]}, index=['영민','정수'])
df_middle
```

### 2) 열 단위 병합
#### 한 번에 두 개의 데이터 프레임만 병합 가능함

#### a) 인덱스가 지정하지 않는 샘플 데이터 프레임 만들기

```python
df_left = DataFrame({'고객번호':[1001,1002,1003,1004],'이름':['철수', '영희', '민철', '미영']})
df_left
```

```python
df_right = DataFrame({'고객번호':[1001,1002,1003,1005],'금액':[10000,20000,15000,5000]})
df_right
```

#### b) 두 데이터 프레임의 공통 컬럼을 기준으로 병합하기

양쪽 데이터 프레임에 모두 존재하는 데이터만 보여준다.
> DataBase의 inner Join과 같은 개념

```python
# 병합 시 제외되는 항목들
# - df_left의 고객번호 1004번 데이터는 df_right의 고객번호 컬럼과 겹치지 않는다.
# - df_right의 고객번호 1005번 데이터는 df_left의 고객번호 컬럼과 겹치지 않는다.
merge(df_left, df_right)
```

#### c) 왼쪽 데이터프레임 기준으로 오른쪽 프레임을 병합하기

`how=left` 파라미터는 왼쪽 데이터프레임의 모든 데이터를 보여준다.
> Database의 Left Outer Join과 같은 개념

```python
merge(df_left, df_right, how='left')
```

#### d) 오른쪽 데이터프레임 기준으로 왼쪽 프레임을 병합하기

`how=right` 파라미터는 오른쪽 데이터프레임의 모든 데이터를 보여준다.
> Database의 Right Outer Join과 같은 개념

```python
merge(df_left, df_right, how='right')
```

#### e) 양쪽 데이터프레임 기준으로 왼쪽 프레임을 병합하기

`how=outer` 파라미터를 추가하면 공통 컬럼의 값이 한쪽만 있어도 데이터를 보여준다.
> Database의 Full Outer Join과 같은 개념

```python
merge(df_left, df_right, how='outer')
```

#### f) 중복되는 데이터가 존재하는 경우 열단위 병합

```python
df_first = DataFrame({'아이디':['hello','world','python','hello'],  '결재금액':[14000,13000,15000,13000]})
df_first
```

```python
df_second = DataFrame({'아이디':['hello','python','python','world'],  '결재금액':[300,500,100,200]})
df_second
```

```python
# on='아이디' : 아이디명을 기준으로 병합한다.
merge(df_first, df_second, on='아이디')
```
