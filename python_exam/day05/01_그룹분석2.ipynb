## #01. 피벗테이블

데이터 열 중 두개의 열을 각각 행, 열로 사용하여 데이터를 재배치하는 것.

### 1) 각 도시별 연도에 따른 인구수 

- 첫번째 파라미터로 지정한 컬럼은 인덱스로 사용됨
- 두번째 파라미터로 지정한 컬럼은 컬럼으로 사용됨
- 세번째 파라미터로 지정한 컬럼은 데이터로 사용됨

```python
pv1 = 인구조사.pivot(index='도시', columns='연도', values='인구')
pv1
```

### 2) 그룹분석 결과를 피벗테이블로 조합하기

각 지역의 연도별 평균 인구수를 피벗 테이블로 생성하자.

####  a) 집계 대상을 추출

```python
df_filter = 인구조사.filter(['지역', '연도', '인구'])
df_filter
```

#### b) 그룹분석

`as_index=False` 파라미터를 사용하여 기준열이 인덱스로 지정되는 것을 방지한다.

```python
df_group = df_filter.groupby(['지역','연도'], as_index=False).sum()
df_group
```

#### c) 피벗 테이블 구성

```python
df_pivot = df_group.pivot(index='지역', columns='연도', values='인구')
df_pivot
```
